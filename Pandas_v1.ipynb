```python
import pandas as pd

mydataset = {
  'cars': ["BMW", "Volvo", "Ford"],
  'passings': [3, 7, 2]
}

myvar = pd.DataFrame(mydataset)

print(myvar)
```

# Panda Series

```python
# without index
a = [1, 7, 2]
myvar = pd.Series(a)
print(myvar)
```

```python
myvar[0]
```

```python
# with index
a = [1, 7, 2]
myvar = pd.Series(a, index = ["x", "y", "z"])
myvar
```

```python
myvar['x']
```

```python
# key/value object as series
calories = {"day1": 420, "day2": 380, "day3": 390}
myvar = pd.Series(calories)
myvar
```

```python
myvar = pd.Series(calories, index = ["day1", "day2"])
myvar
```

# Grouping and Aggregation

```python
import pandas as pd

data = {
    'Store': ['North', 'South', 'North', 'West', 'South', 'West'],
    'Category': ['Electronics', 'Clothing', 'Clothing', 'Electronics', 'Electronics', 'Clothing'],
    'Sales': [1000, 500, 800, 1200, 900, 600],
    'Quantity': [5, 3, 4, 10, 15, 6]
}
df = pd.DataFrame(data)
df.groupby('Store').sum()
```

```python
df.groupby('Store')['Sales'].sum()
```

```python
df.groupby('Store')['Sales'].size()
```
