# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,4,5]
y = [10,20,30,40,50]

plt.plot(x,y)
plt.show()
```

## Parts of a Figure

```python
plt.plot(x,y,label="Revenue")
plt.title("Company Revenue")
plt.xlabel("Month")
plt.ylabel("Income")
plt.grid(True)
plt.legend();
```

## Scatter plot

```python
plt.scatter(x,y, label="Revenue", zorder=2, color="red", s=40); # s = size
plt.title("Company Revenue")
plt.xlabel("Month")
plt.ylabel("Income")
plt.grid(True, zorder=0)
plt.legend();
```

## Histogram

```python
plt.hist(y)
plt.title("Histogram of y")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

## Bar Chart

```python
plt.bar(x,y, label="Revenue")
plt.title("Company Revenue")
plt.xlabel("Month")
plt.ylabel("Income");
```

## Pie Chart

```python
plt.pie(y, labels=x, autopct="%1.1f%%")
plt.title("Company Revenue")
plt.show()
```

## Day 7: Subplots

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
ax[0].scatter(x,y, label="Revenue", zorder=2, color="red", s=40)
ax[0].set_title("Company Revenue (Scatter Plot)")
ax[0].set_xlabel("Month (x-axis)")
ax[0].set_ylabel("Income (y-axis)")
ax[0].grid(True)
ax[0].legend()

ax[1].plot(x,y, label="Revenue")
ax[1].set_title("Company Revenue (Line Plot)")
ax[1].set_xlabel("Month (x-axis)")
ax[1].set_ylabel("Income (y-axis)")
ax[1].grid(True)
ax[1].legend()

plt.tight_layout()
plt.show()
```
