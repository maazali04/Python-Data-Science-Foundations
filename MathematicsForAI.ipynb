# Level 0: Foundation

## 1. Numbers (The Building Blocks)
  - **Natural ($\mathbb{N}$)**: Counting numbers starting from 1 ($1, 2, 3, \dots$).
  - **Integers ($\mathbb{Z}$)**: Whole numbers, including zero and negatives ($\dots, -2, -1, 0, 1, 2, \dots$).
  - **Rational ($\mathbb{Q}$)**: Numbers expressible as a fraction $\frac{a}{b}$ (e.g., $0.5, \frac{3}{4}$). They terminate or repeat.
  - **Irrational ($\mathbb{I}$)**: Numbers that cannot be written as fractions (e.g., $\pi, \sqrt{2}$). Decimals go on forever without repeating.
  - **Real ($\mathbb{R}$)**: The entire continuous spectrum of all rational and irrational numbers combined (any point on a number line).
  - **Complex ($\mathbb{C}$)**: Numbers containing a real part and an imaginary part, written as $a + bi$ (where $i = \sqrt{-1}$).
  
### Why We Care in AI/ML
  - **Real Numbers ($\mathbb{R}$)**: The absolute foundation of modern Machine Learning. Weights, biases, inputs, and gradients in neural networks are continuous real numbers. They allow algorithms to perform tiny calculus adjustments (gradient descent) to minimize error.
  - **Complex Numbers ($\mathbb{C}$)**: Vital for advanced processing signal domains. Algorithms like the Fourier Transform rely on complex numbers to decompose audio waves, frequencies, or medical images (like MRIs) into mathematical patterns before feeding them into an AI model.

## 2. Numbers (The Building Blocks)

- **Variables ($x, y, z$)**: Symbols (usually letters) that act as placeholders for values that can change or are unknown.
- **Constants ($5, -3, \pi$)**: Fixed values that never change.
- **Expressions ($3x + 5$)**: A mathematical phrase combining variables, constants, and operators ($+$, $-$, $\times$, $\div$). It does not have an equals sign ($=$) and cannot be solved, only simplified.
- **Equations ($3x + 5 = 20$)**: A statement that two expressions are equal. It contains an equals sign ($=$) and can be solved to find the specific value of the variable.
- **Inequalities ($3x + 5 > 20$)**: Statements showing that two expressions are not equal, using relation symbols like $>$, $<$, $\geq$, or $\leq$. Instead of one answer, they result in a range of possible answers.
### Why We Care in AI/ML
- Variables & Constants: In a machine learning model (like $y = wx + b$), the input data ($x$) is the variable, while the weights ($w$) and biases ($b$) are the parameters the computer adjusts. Once trained, those weights become fixed constants used to make predictions ($y$).
- Equations: Training a model means solving a massive system of equations. Every time an AI makes a prediction, it evaluates an algebraic equation.
- Inequalities: Crucial for Classification models and Activation Functions. For example, a model uses inequalities to create a decision boundary: If probability $> 0.5$, classify as Spam; else, classify as Not Spam.

## 3. Functions (The Transformers)
- $f(x)$: The standard notation for a function. Read as "f of x". It is a mathematical machine that takes an input, processes it through a specific rule, and gives an output.
- Input ($x$): The independent variable you feed into the function.
- Output ($y$ or $f(x)$): The dependent variable that comes out of the function after the math is applied.
- Domain: The set of all possible input values ($x$) that the function can safely accept without breaking (e.g., without dividing by zero or taking the square root of a negative number).
- Range: The set of all possible output values ($y$) that the function can actually produce.
### Why We Care in AI/ML
- Machine Learning is a Function: At its absolute core, an entire Machine Learning model is just a giant, complex mathematical function:$$y = f(x)$$Your input features (like house size) are passed into the trained function $f$, and the function calculates the output prediction $y$ (the house price).
- Activation Functions: In Deep Learning and Neural Networks, specific functions dictate how neurons fire. For example:
  - The Sigmoid function maps any real number input down to a strict Range of $(0, 1)$. This range is perfect because it translates raw numbers into clean probabilities (e.g., $0.85 = 85\%$ chance of being a cat).
  - The ReLU function sets a Domain-specific rule: If the input is negative, output exactly 0. If it's positive, pass it through as-is.

### 4. Graphs (Visualizing the Behavior)
- Linear: A straight-line relationship ($y = mx + b$). As $x$ increases, $y$ increases or decreases at a perfectly constant rate.
- Quadratic: A parabola or U-shaped curve ($y = ax^2 + bx + c$). It possesses a single distinct minimum or maximum point where the direction changes.
- Exponential: A curve that explodes upward or decays downward rapidly ($y = a^x$). The rate of change grows faster and faster as $x$ increases.
- Logarithmic: The mathematical inverse of exponential ($y = \log(x)$). It climbs quickly at first, but then flattens out, growing slower and slower as $x$ increases.
#### Why We Care in AI/ML
- Loss Functions (Quadratic): When measuring how wrong a model's predictions are, we use functions like Mean Squared Error (MSE), which are quadratic. Because it forms a U-shape, the absolute bottom of the curve represents the lowest possible error. The model uses optimization techniques to "roll down" this curve to find that perfect minimum point.
- Activation Functions (Exponential & Logarithmic): * The Sigmoid and Softmax activation functions use exponential math ($e^x$) to smoothly squash raw numbers into a probability distribution.
  - The Binary Cross-Entropy Loss (used to measure error in classification models) relies entirely on logarithmic graphs to heavily penalize the model if it is confident but completely wrong.

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(2,2)
fig.set_figheight(6)
fig.set_figwidth(6)

x_common = np.linspace(-10,10,100)
x_log = np.linspace(0.1, 10, 100) # x for log must be positive

y = 3*x_common + 5
y2 = 3*x_common**2 + 5
y3 = 2**x_common
y4 = np.log(x_log)

ax[0][0].plot(x_common, y)
ax[0][0].set_title('Linear: y = 3x + 5')
ax[0][0].grid()
ax[0][1].plot(x_common, y2)
ax[0][1].set_title('Quadratic: y = 3x^2 + 5')
ax[0][1].grid()
ax[1][0].plot(x_common, y3)
ax[1][0].set_title('Exponential: y = 2^x')
ax[1][0].grid()
ax[1][1].plot(x_log, y4)
ax[1][1].set_title('Logarithmic: y = log(x)')
ax[1][1].grid()

plt.tight_layout()
plt.show()
```

## Scalars (Single Values)
- Definition: A scalar is a single numerical value (a 0-dimensional tensor) with magnitude but no direction.
- Addition & Subtraction ($a + b$, $a - b$): Shifting a single value up or down along a number line.
- Multiplication & Division ($a \cdot b$, $\frac{a}{b}$): Scaling a single value up or down by a specific factor.
- Powers ($a^n$): Multiplying a scalar by itself $n$ times. Governs rapid growth or non-linear scaling.
- Roots ($\sqrt[n]{a}$): The inverse of a power. Finding a value that, multiplied by itself $n$ times, equals the scalar.
### Why We Care in AI/ML
- Single Metrics (Loss Value): When your model calculates its error across an entire dataset, it compresses all that data down into a single scalar value (e.g., $\text{Loss} = 0.24$). This single number tells you exactly how well or poorly your model is performing overall.
- Hyperparameters (Learning Rate $\alpha$): The learning rate is a critical scalar (typically a tiny decimal like $0.01$ or $0.001$). During training, the model multiplies this tiny scalar by the calculated gradient vectors to control how large or small of a step the algorithm takes when updating its weights.
- Scalar Broadcasting (NumPy/Pandas): When you multiply an entire multi-dimensional NumPy array or Pandas column by a single scalar, NumPy automatically "broadcasts" that scalar across every individual element in the structure without requiring slow for loops.

```python
# Native Python scalars
native_a = 5
native_b = 10

# Numpy Scalars
import numpy as np
np_a = np.array(5)
np_b = np.array(10)

print("Addition")
print(native_a + native_b)
print(np_a + np_b)

print("Subtraction")
print(native_a - native_b)
print(np_a - np_b)

print("Multiplication")
print(native_a * native_b)
print(np_a * np_b)

print("Division")
print(native_a / native_b)
print(np_a / np_b)

print("Power")
print(native_a**3)
print(np_a**3)

print("Root")
print(native_a**(1/3))
print(np_a**(1/3))
```

## Vectors (Data with Direction)
- Vector: An ordered list of numbers representing a point in space. Geometrically, it can be visualized as an arrow with both magnitude (length) and direction.
- Vector Creation: Written as a column or row matrix, denoted by a lowercase bold letter or an arrow: $\vec{v} = [x_1, x_2, x_3]$. In Python, you create this using a 1D NumPy array: v = np.array([3, 4]).
- Magnitude ($\|\vec{v}\|$): The total geometric length of the vector. Calculated using the Pythagorean theorem formula:$$\|\vec{v}\| = \sqrt{x_1^2 + x_2^2 + \dots + x_n^2}$$
- Direction: The spatial orientation or angle of the vector arrow relative to the coordinate axes.
- Unit Vector ($\hat{v}$): A vector with a magnitude of exactly 1. Created by dividing a vector by its own magnitude ($\hat{v} = \frac{\vec{v}}{\|\vec{v}\|}$). This process is called normalization.

### Vector Operations
- Addition & Subtraction: Done element-wise (e.g., $[1, 2] + [3, 4] = [4, 6]$). Geometrically, vector addition is stacking arrows tip-to-tail.
- Scalar Multiplication: Scalar multiplication scales a vector's length, and if the scalar is negative, it flips the vector's direction to point the opposite way. (e.g., $2 \cdot [3, 4] = [6, 8]$).

###  Why We Care in AI/ML
- Feature Vectors: This is how datasets are structured. Every row in your Pandas DataFrame is turned into a feature vector before hitting a Scikit-Learn algorithm. For instance, a house vector could look like $\vec{x} = [\text{sq\_ft}, \text{bedrooms}, \text{price}]$.
- Embeddings & Word Vectors: Text cannot be read by algorithms. We use models like Word2Vec to translate human words into long vectors of real numbers (e.g., 300 dimensions).
  - Because these vectors capture semantic meaning, they can do Vector Arithmetic! A famous example is:$$\vec{\text{King}} - \vec{\text{Man}} + \vec{\text{Woman}} = \vec{\text{Queen}}$$
- Similarity Metrics: To build a recommendation system, the computer calculates the angle between two word/item vectors using Cosine Similarity. If the vectors point in nearly the exact same direction (small angle), the items are highly similar.

### Vector Norms
- A Norm is just a function that takes a vector and returns a single positive scalar representing the "size" or "length" of that vector.

1.  $L_0$ Norm
  - What it does: It is not a true mathematical norm, but it counts the number of non-zero elements in a vector.
  - Math: If $\vec{v} = [3, 0, -1, 5, 0]$, then $\|\vec{v}\|_0 = 3$ (because there are 3 non-zero numbers).

2. $L_1$ Norm (Manhattan / Taxicab Distance)
  - What it does: Calculates the sum of the absolute values of the vector components. It measures the distance if you could only travel along grid lines (like a taxi driving through Manhattan streets).
  - Math formula:$$\|\vec{v}\|_1 = \sum_{i=1}^{n} |x_i| = |x_1| + |x_2| + \dots + |x_n|$$.

3. $L_2$ Norm (Euclidean Norm)
  - What it does: Calculates the standard straight-line distance from the origin to the vector coordinate (the classic Pythagorean distance). This is the default norm when people say "length".
  - Math formula: $$\|\vec{v}\|_2 = \sqrt{\sum_{i=1}^{n} x_i^2} = \sqrt{x_1^2 + x_2^2 + \dots + x_n^2}$$

4. Infinity Norm ($L_\infty$ Norm)
  - What it does: Returns the absolute maximum value among all the elements in the vector. It completely ignores all other smaller components.
  - Math formula:$$\|\vec{v}\|_\infty = \max(|x_1|, |x_2|, \dots, |x_n|)$$
  
#### Why We Care in AI/ML
1. Regularization (Preventing Overfitting)
    - When a machine learning model overfits, its weights ($w$) blow up and become massive numbers. We use norms to punish the model for having large weights by adding them to the Loss function:
      - Lasso Regression ($L_1$ Regularization): Adds the $L_1$ norm of the weights to the loss. Because of its sharp geometric nature, it drives less important weights exactly to 0, automatically performing feature selection.
      - Ridge Regression ($L_2$ Regularization): Adds the $L_2$ norm of the weights to the loss. It shrinks all weights down to tiny decimals close to zero, preventing any single feature from dominating the model.2.
  
2. Distance Calculations (Similarity Metrics)
    - $L_2$ Distance is the foundation of algorithms like K-Nearest Neighbors (KNN) and K-Means Clustering to check how close data points are to each other in space.
    - $L_1$ Distance is preferred over $L_2$ in incredibly high-dimensional data fields because $L_2$ becomes distorted when dealing with the Curse of Dimensionality.

3. Optimization & Deep Learning
    - Gradient Clipping: In Deep Learning, gradients can sometimes explode and become massive, causing the model to break. We calculate the $L_2$ norm of the gradient vector; if it exceeds a certain threshold scalar, we scale it down to keep training stable.

```python
import numpy as np

v = np.array([3, -4, 0, 5])

# L0 Norm
l0 = np.count_nonzero(v)

# L1 Norm
l1 = np.linalg.norm(v, ord=1)

# L2 Norm
l2 = np.linalg.norm(v, ord=2)

# Infinity Norm
l_inf = np.linalg.norm(v, ord=np.inf)

print(f"L0 Norm: {l0}")
print(f"L1 Norm: {l1}")
print(f"L2 Norm: {l2}")
print(f"Linf Norm: {l_inf}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a dense grid of points between -2 and 2
x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(x, y)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
titles = ["L1 Norm (Manhattan)", "L2 Norm (Euclidean)", "L∞ Norm (Max Value)"]

# 1. Calculate the norm grids
Z_L1 = np.abs(X) + np.abs(Y)
Z_L2 = np.sqrt(X**2 + Y**2)
Z_Linf = np.maximum(np.abs(X), np.abs(Y))

grids = [Z_L1, Z_L2, Z_Linf]
colors = ["red", "blue", "green"]

for i, ax in enumerate(axes):
    ax.axhline(0, color='black', linewidth=1)
    ax.axvline(0, color='black', linewidth=1)
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.grid(True, linestyle=":", alpha=0.5)
    ax.set_title(titles[i], fontsize=12, fontweight='bold')
    ax.set_aspect('equal')

    # Draw the Unit Circle contour (where Norm = 1)
    contour = ax.contour(X, Y, grids[i], levels=[1.0], colors=colors[i], linewidths=3)
    ax.clabel(contour, fmt={1.0: 'Norm = 1'}, inline=True, fontsize=10)

plt.tight_layout()
plt.show()
```

### Dot Product (The Similarity Engine)
- The Dot Product takes two vectors of the same length and returns a single scalar value. You can calculate it in two ways:
  - Algebraic (Element-wise): Multiply corresponding elements and add them up.$$\vec{A} \cdot \vec{B} = (A_1 \cdot B_1) + (A_2 \cdot B_2) + \dots + (A_n \cdot B_n)$$
  - Geometric (Angle-based):$$\vec{A} \cdot \vec{B} = \|\vec{A}\| \|\vec{B}\| \cos(\theta)$$(Where $\theta$ is the angle between the two vectors).
  - Note: Shows alignment. It answers: "Are these two vectors tracking together, completely independent, or working against each other?
  
#### Why We Care in AI/ML
  - The Ultimate Similarity Metric: Look at the geometric formula. If two vectors point in the exact same direction, $\theta = 0^\circ$ and $\cos(0) = 1$ (the dot product is maximized). If they are perpendicular ($90^\circ$), $\cos(90) = 0$ (the dot product is zero). This makes it the foundation of Cosine Similarity used in recommendation systems.
  - Neural Networks ($W \cdot X$): Every single neuron in a neural network calculates a dot product. It takes a input feature vector ($\vec{X}$) and multiplies it by its internal weight vector ($\vec{W}$), plus a bias.
  - Transformers & Attention Mechanisms: Modern GenAI (like ChatGPT or Gemini) relies on Large Language Models. Under the hood, the Self-Attention mechanism calculates the dot product between "Query" vectors and "Key" vectors to see which words in a sentence relate most heavily to each other.

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([3, 4])
B = np.array([5, 1])

dot_prod = np.dot(A, B) # 3*5 + 4*1 = 19

# Formula for projection vector of A onto B: (A . B) / ||B||^2 * B
norm_B_sq = np.sum(B**2)
proj_A_on_B = (dot_prod / norm_B_sq) * B

plt.figure(figsize=(8, 8))
plt.quiver(0, 0, A[0], A[1], angles='xy', scale_units='xy', scale=1,
         color="blue", zorder=3, width=0.012, label=f"Vector A {list(A)}")
plt.quiver(0, 0, B[0], B[1], angles='xy', scale_units='xy', scale=1,
           color="red", zorder=3, width=0.012, label=f"Vector B {list(B)}")
plt.quiver(0, 0, proj_A_on_B[0], proj_A_on_B[1], angles='xy', scale_units='xy', scale=1,
           color="purple", zorder=2, width=0.015, alpha=0.6, label="Shadow of A on B")
plt.plot([A[0], proj_A_on_B[0]], [A[1], proj_A_on_B[1]],
         color="black", linestyle=":", linewidth=2, label="Perpendicular Projection Line")
text_info = f"A · B = {dot_prod}\n||B|| = {np.linalg.norm(B):.2f}\nShadow Length = {np.linalg.norm(proj_A_on_B):.2f}"
plt.text(0.5, 3.5, text_info, fontsize=11, fontweight="bold",
         bbox=dict(facecolor='white', alpha=0.8, edgecolor='gray'))

plt.title("Geometric Meaning of Dot Product (Vector Projection)", fontsize=13, pad=15)
plt.xlim(-1, 6)
plt.ylim(-1, 6)
plt.axhline(y=0, color='k', zorder=1)
plt.axvline(x=0, color='k', zorder=1)
plt.xticks(range(7))
plt.yticks(range(7))
plt.grid(True, linestyle=":", alpha=0.5)
plt.gca().set_aspect('equal')
plt.legend(loc="upper right")
plt.show()
```

### Cross Product (The Perpendicular Vector Creator)
- Unlike the Dot Product, the Cross Product ($\vec{A} \times \vec{B}$) is only defined in 3D space and returns a brand new vector, not a scalar.
  - The Intuition: The resulting vector points in a direction that is perfectly perpendicular ($90^\circ$) to both original vectors. You can find its direction using the "Right-Hand Rule.
  - "The Magnitude: The length of the resulting cross-product vector is exactly equal to the area of the parallelogram formed by the two original vectors.
### Why We Care in AI/ML
  - Rarely Used in General ML: You will almost never see a cross product in standard tabular machine learning, deep learning, or NLP.
  - Computer Vision & Robotics: It is occasionally used in 3D computer vision, self-driving car spatial awareness, or robotics. For example, if an AI needs to calculate a Normal Vector to detect the flat surface of a wall or a floor from 3D camera point-clouds, it uses the cross product to find that perpendicular direction.

```python
import numpy as np

A = np.array([1, 2, 3])
B = np.array([4, 5, 6])

np.cross(A,B)
```

## Matrices (The Engines of AI)
- A matrix is a 2D grid of numbers organized into rows and columns. In code, you treat this as a 2D NumPy array (ndim == 2).
- Rows ($m$): The horizontal lines of numbers.
- Columns ($n$): The vertical lines of numbers.
- Shape ($m \times n$): Always written as (Rows $\times$ Columns). A matrix with 3 rows and 2 columns has a shape of $3 \times 2$.
###  Basic Operations
  - Addition & Subtraction: Done element-wise. The two matrices must have the exact same shape.
  - Scalar Multiplication: Multiplying every individual element in the matrix by a single scalar value.

```python
import numpy as np

matrix_A = np.array([[1, 2],
                     [5, 6]])

matrix_B = np.array([[10, 20],
                     [50, 60]])

scalar_value = 2

matrix_sum = matrix_A + matrix_B
matrix_diff = matrix_B - matrix_A
matrix_scaled = matrix_A * scalar_value

print("Matrix A:\n", matrix_A)
print("Matrix B:\n", matrix_B)
print("\n1. Addition (A + B):\n", matrix_sum)
print("\n2. Subtraction (B - A):\n", matrix_diff)
print(f"\n3. Scalar Multiplication (A * {scalar_value}):\n", matrix_scaled)
```

### Matrix Multiplication (Dot Product of Matrices)
- Matrix multiplication ($A \times B$) is not element-wise. Instead, it is a series of dot products matching the rows of the first matrix with the columns of the second matrix.
- The Shape Rule: To multiply matrix $A$ by matrix $B$, the columns of $A$ must equal the rows of $B$.
$$\text{Shape: } (m \times \mathbf{n}) \times (\mathbf{n} \times k) \longrightarrow (m \times k)$$
- What the result shows: It represents a composite linear transformation. It maps or projects features from one dimensional space to another.

#### AI Use:
- The absolute core of AI. Every layer in a neural network calculates $Y = \sigma(W \cdot X + b)$, where $W$ is a massive weight matrix and $X$ is your input data matrix. Every transformer and LLM uses this to process tokens simultaneously.

```python
import numpy as np
matrix_a = np.array([[1,2],[3,4]])
matrix_b = np.array([[5,6],[7,8]])

matrix_a @ matrix_b
```

### Matrix Transpose ($A^T$)
- What it does: Flips a matrix over its diagonal, swapping its rows with its columns. If matrix $A$ has a shape of $3 \times 2$, then $A^T$ will have a shape of $2 \times 3$.
- What the result shows: It re-orients the coordinate space, changing feature rows into feature columns.

#### AI Use:
- Backpropagation. When a neural network calculates errors to update its weights, the dimensions of the error vectors don't naturally line up with the weight matrices. We transpose the matrices to flip their shapes so matrix multiplication becomes mathematically possible during gradient descent.

```python
matrix_a = np.array([[1,2],[3,4]])
matrix_a.T
```

### Identity Matrix ($I$)
- What it does: The matrix equivalent of the number 1. It is a square matrix with $1$s on the main diagonal and $0$s everywhere else.
- What the result shows: Multiplying any matrix by $I$ leaves the original matrix completely unchanged ($A \cdot I = A$). It represents a transformation that does absolutely nothing to your spatial coordinates.
#### AI Use:
- Acts as the baseline or starting point for linear transformations and structural network architectures (like Residual Connections/ResNets, which pass an identity mapping of data forward).

```python
np.eye(2)
```

### Determinants ($\det(A)$ or $|A|$)
- What it does: A function that takes a square matrix and outputs a single scalar number.
- What the result shows: It measures how much space scales or expands during a transformation.
  - If $\det(A) = 2$, the transformation doubles the spatial area.
  - If $\det(A) = 0$, the transformation completely squashes space down into a flat line or single point, meaning information is permanently lost, and the matrix cannot be inverted.
#### AI Use:
- Checking Matrix Invertibility. If $\det(A) = 0$, the matrix is singular and cannot be inverted, which can completely crash traditional statistical algorithms trying to calculate exact solutions.

```python
matrix_a = np.array([[1,2],[3,4]])
np.linalg.det(matrix_a)
```

### Inverse Matrix ($A^{-1}$)
- What it does: The matrix equivalent of division ($x^{-1} = \frac{1}{x}$). It undoes whatever the original matrix did.$$A \cdot A^{-1} = I$$
- What the result shows: It perfectly reverses or unwinds a spatial transformation.

#### AI Use:
- Linear Regression (Normal Equation). Traditional closed-form linear regression finds the absolute best-fitting line in a single mathematical step using the formula: $\theta = (X^T X)^{-1} X^T y$. It explicitly requires calculating an inverse matrix.

```python
matrix_a = np.array([[1,2],[3,4]])
np.linalg.inv(matrix_a)
```

### Matrix Rank
- What it does: The number of linearly independent rows or columns in a matrix.
- What the result shows: It indicates the true informational dimensionality of your data. If you have a $10 \times 10$ matrix, but 3 columns are just duplicates or multiples of other columns, the matrix is Rank deficient (Rank = 7).

#### AI Use:
- Feature Redundancy. If your dataset has low rank, it means you have highly redundant columns (multicollinearity). AI developers look at rank to see if they can use Dimensionality Reduction (like PCA) to compress the dataset without losing any actual underlying data signals.

### Linear Transformations
- What it is: A function that takes a vector input, moves it through a coordinate space using matrix multiplication ($A\vec{v}$), and outputs a new shifted vector.

- What the result shows: It shows how a matrix stretches, rotates, shears, or squashes a geometric space.

#### AI Use:

### Eigenvectors & Eigenvalues
- When a matrix transforms space, it usually knocks vectors off their original path, rotating them. However, a few special vectors do not change their direction; they only stretch or shrink.
  - Eigenvector ($\vec{v}$): A special vector that stays on its exact same span (line) during a linear transformation. Its direction is completely unchanged.
  - Eigenvalue ($\lambda$): The scalar factor by which that eigenvector stretches, shrinks, or flips.$$\mathbf{A}\vec{v} = \lambda\vec{v}$$
  $$det (\mathbf{A} - \lambda \mathbf{I}) = 0$$
  - What the result shows: Eigenvectors point along the axes of maximum variance or structural symmetry within a dataset. Eigenvalues tell you how much data variance or power is concentrated along that specific axis.

#### AI Use:
- Principal Component Analysis (PCA): PCA finds the eigenvectors of a dataset's covariance matrix. It drops the eigenvectors with tiny eigenvalues (low variance) and keeps the ones with massive eigenvalues, compressing thousands of features into a few core dimensions.
  - PageRank Algorithm: Google's original search algorithm uses the dominant eigenvector of a web-link matrix to rank the importance of websites.

#### Singular Value Decomposition (SVD)
- What it is: The ultimate factorization tool. It states that any matrix $A$ (not just square ones) can be broken down into three distinct, simpler matrices:$$\mathbf{A} = \mathbf{U} \mathbf{\Sigma} \mathbf{V}^T$$

  - $\Sigma$: Singular values (Scaling matrix/Diagonal).
  - $V^T$: Right singular vectors (orthogonal vector) (Rotation matrix).
- What the result shows: SVD systematically breaks any complex spatial transformation down into three clean geometric steps: Rotate $\rightarrow$ Scale $\rightarrow$ Rotate. It ranks the structural information of a matrix from most important to least important.

#### AI Use:

## Limits
- A limit evaluates the value that a function $f(x)$ approaches as the independent variable $x$ gets infinitely close to a target value $c$, regardless of the function's actual value at $c$.
- Mathematically: $$\lim_{x \to c} f(x) = L$$
- A two-sided limit exists if and only if the left-hand limit (approaching from values $x < c$) and the right-hand limit (approaching from values $x > c$) are equal.
 $$\text{Existence Condition: } \lim_{x \to c} f(x) = L \iff \lim_{x \to c^-} f(x) = \lim_{x \to c^+} f(x) = L$$
 If $L_1 \neq L_2$ (such as at a step function's jump discontinuity), the limit Does Not Exist (DNE).

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp


x = sp.Symbol('x')
f = sp.Abs(x - 2) / (x - 2)
target_c = 2

left_limit = sp.limit(f, x, target_c, dir='-')
right_limit = sp.limit(f, x, target_c, dir='+')

print(f"--- Analytical Mathematical Report ---")
print(f"Left-Hand Limit (approaching {target_c}-): {left_limit}")
print(f"Right-Hand Limit (approaching {target_c}+): {right_limit}")

if left_limit != right_limit:
    limit_status = "Does Not Exist (DNE)"
    print(f"Result: The two-sided limit {limit_status} due to a jump discontinuity.")
else:
    limit_status = str(left_limit)
    print(f"Result: The two-sided limit exists and equals {limit_status}.")


x1 = np.linspace(0, 1.99, 200)
x2 = np.linspace(2.01, 4, 200)

y1 = np.abs(x1 - 2) / (x1 - 2)
y2 = np.abs(x2 - 2) / (x2 - 2)

plt.figure(figsize=(7, 4.5))

plt.plot(x1, y1, 'b-', linewidth=2, label=r'$f(x) = \frac{|x-2|}{x-2}$')
plt.plot(x2, y2, 'b-', linewidth=2)

plt.plot(2, -1, 'bo', markerfacecolor='white', markersize=8, label='Excluded Point')
plt.plot(2, 1, 'bo', markersize=8, label='Target Value')

plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.axvline(2, color='red', linestyle='--', linewidth=1, label=r'Target $x \to 2$', zorder=1)

plt.title(f'Limit Analysis Plotting: Limit Status = {limit_status}', fontsize=12)
plt.xlabel('$x$ Coordinate Space', fontsize=10)
plt.ylabel('$f(x)$ Value Space', fontsize=10)
plt.ylim(-2, 2)
plt.xlim(0, 4)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(loc='best')

plt.show()
```

## Continuity

- A function $f(x)$ is continuous at a specific point if its graph has no holes, jumps, or vertical asymptotes at that point. Formally, continuity bridges the behavior of a function's limit with its actual output value.

### The Three Conditions for Continuity
- The function is defined at $c$.
- The limit exists at $c$.
- The limit equals the function value.

If any of these three conditions fail, the function is discontinuous at $x = c$

### Types of Discontinuity
1. Removable Discontinuity (Hole)
  - The limit $\lim_{x \to c} f(x)$ exists, but $\lim_{x \to c} f(x) \neq f(c)$ (either $f(c)$ is completely undefined or sits at a different point).

2. Jump Discontinuity
  - The one-sided limits both exist as finite numbers, but they are not equal.

3. Infinite Discontinuity (Asymptote)
  - One or both of the one-sided limits approach positive or negative infinity ($\pm\infty$).

### Operational Relevance in Machine Learning:
Continuity is a strict prerequisite for optimization.
  - Differentiability Requires Continuity: For an activation function or a loss function $J(\theta)$ to be differentiable (meaning you can calculate its gradients via backpropagation), it must be continuous. If a function contains a jump or infinite discontinuity, the derivative cannot be computed at that point, causing gradient updates to fail.
  - The ReLU exception ($x=0$): The Rectified Linear Unit ($\text{ReLU}(x) = \max(0, x)$) is completely continuous everywhere, even at the sharp corner of $x=0$. While it is technically non-differentiable exactly at $0$ due to the sharp turn, its continuity allows optimization packages to easily bypass this using subgradients.

```python
import numpy as np
import matplotlib.pyplot as plt


x_clean = np.linspace(-3, 3, 200)
x_left = np.linspace(-3, 0.99, 100)
x_right = np.linspace(1.01, 3, 100)

y1 = x_clean ** 2

y2_left = ((x_left - 1) * (x_left + 2)) / (x_left - 1)
y2_right = ((x_right - 1) * (x_right + 2)) / (x_right - 1)

y3_left = x_left ** 2
y3_right = x_right + 2

y4_left = (x_left + 1) / (x_left - 1)
y4_right = (x_right + 1) / (x_right - 1)

fig, ax = plt.subplots(2, 2, figsize=(10, 7.5))


ax[0, 0].plot(x_clean, y1, color='blue')
ax[0, 0].set_title("1. Continuous Function")
ax[0, 0].set_ylim(-4, 6)

# -------------------------------------------------------------

ax[0, 1].plot(x_left, y2_left, color='blue')
ax[0, 1].plot(x_right, y2_right, color='blue')
ax[0, 1].plot(1, 3, marker='o', markerfacecolor='white', markeredgecolor='blue', markersize=6)
ax[0, 1].plot(1, 1, marker='o', color='red', markersize=5)
ax[0, 1].set_title("2. Removable Discontinuity (Hole)")
ax[0, 1].set_ylim(-4, 6)

ax[1, 0].plot(x_left, y3_left, color='blue')
ax[1, 0].plot(x_right, y3_right, color='blue')
ax[1, 0].plot(1, 1, marker='o', markerfacecolor='white', markeredgecolor='blue', markersize=6)
ax[1, 0].plot(1, 3, marker='o', color='blue', markersize=5)
ax[1, 0].set_title("3. Jump Discontinuity")
ax[1, 0].set_ylim(-4, 6)

ax[1, 1].plot(x_left, y4_left, color='blue')
ax[1, 1].plot(x_right, y4_right, color='blue')
ax[1, 1].axvline(x=1, color='red', linestyle='--')
ax[1, 1].set_title("4. Infinite Discontinuity")
ax[1, 1].set_ylim(-10, 10)

for a in ax.flat:
    a.grid(True, linestyle=':', alpha=0.6)
    a.axhline(0, color='black', linewidth=0.5)
    a.axvline(0, color='black', linewidth=0.5)
    a.set_xlabel("x")
    a.set_ylabel("y")

plt.tight_layout()
plt.show()
```

## Derivatives
  - A derivative represents the instantaneous rate of change of a function $f(x)$ with respect to $x$. Formally, it is defined as the limit of a difference quotient as the interval approaches zero:$$\frac{df}{dx} = f'(x) = \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x}$$
  
###  The Four Core Derivative Rules
1. The Power Rule:
  - Used when the variable $x$ is raised to a constant power $n$.$$\frac{d}{dx}[x^n] = n \cdot x^{n-1}$$
  - Mathematical Example: If $f(x) = x^4$, then $f'(x) = 4x^3$.

2. The Product Rule:
  - Used when a function is the product of two differentiable sub-functions, $u(x)$ and $v(x)$.$$\frac{d}{dx}[u \cdot v] = u \cdot \frac{dv}{dx} + v \cdot \frac{du}{dx}$$
  - Mathematical Example: If $f(x) = x^2 \cdot \sin(x)$:
    - $u = x^2 \implies du = 2x$
    - $v = \sin(x) \implies dv = \cos(x)$
    - $f'(x) = x^2\cos(x) + 2x\sin(x)$
  
3. The Quotient Rule:
  - Used when a function is the division of two differentiable sub-functions, $u(x)$ divided by $v(x)$.$$\frac{d}{dx}\left[\frac{u}{v}\right] = \frac{v \cdot \frac{du}{dx} - u \cdot \frac{dv}{dx}}{v^2}$$
  - Mathematical Example: If $f(x) = \frac{\ln(x)}{x}$:
    - $u = \ln(x) \implies du = \frac{1}{x}$
    - $v = x \implies dv = 1$
    - $f'(x) = \frac{x(\frac{1}{x}) - \ln(x)(1)}{x^2} = \frac{1 - \ln(x)}{x^2}$
  
4. The Chain Rule:
  - Used for composite functions $f(g(x))$, where one function is nested inside another. This rule is the mathematical core of Backpropagation in neural networks.$$\frac{d}{dx}[f(g(x))] = f'(g(x)) \cdot g'(x)$$
  - Mathematical Example: If $f(x) = (3x^2 + 1)^5$:
    - Outer function: $u^5 \implies 5u^4$
    - Inner function: $3x^2 + 1 \implies 6x$
    - $f'(x) = 5(3x^2 + 1)^4 \cdot 6x = 30x(3x^2 + 1)^4$

###  Operational Relevance in Machine Learning
  - Every time an optimizer calculates gradients, it applies these rules. For example, computing the derivative of the Sigmoid Activation Function $\sigma(z) = \frac{1}{1 + e^{-z}}$ requires applying the Chain Rule and the Quotient Rule, resolving to the clean property: $\sigma'(z) = \sigma(z)(1 - \sigma(z))$.

```python
import sympy as sp

x = sp.Symbol('x')

f_power = x**4
df_power = sp.diff(f_power, x)
print(f"Power Rule: {df_power}")

f_product = (x**2) * sp.sin(x)
df_product = sp.diff(f_product, x)
print(f"Product Rule: {df_product}")

f_quotient = sp.log(x) / x
df_quotient = sp.diff(f_quotient, x)
print(f"Quotient Rule: {df_quotient}")

f_chain = (3 * x**2 + 1) ** 5
df_chain = sp.diff(f_chain, x)
print(f"Chain Rule: {df_chain}")
```

## Partial Derivatives and Gradient
## Partial Derivatives:($\frac{\partial f}{\partial x}$)
  - A standard derivative looks at a function with only one input variable. A partial derivative takes a function with multiple inputs and measures how the output changes when you alter only one variable while freezing all the others completely constant.
  
  -  The Mathematical Rule
      - To calculate a partial derivative with respect to variable $x$ (written as $\frac{\partial f}{\partial x}$):
        - Treat every instance of $x$ as your normal variable.
        - Treat all other variables ($y, z, w$, etc.) exactly like fixed numbers (constants).
        
      - Calculation Example:Let our function be:$$f(x, y) = 4x^3y^2 + 7x + 2y^4$$
      
        - Find $\frac{\partial f}{\partial x}$ (Treat $y$ as a constant number):$$\frac{\partial f}{\partial x} = (3 \cdot 4)x^2 \cdot y^2 + 7 + 0 = 12x^2y^2 + 7$$(Note that $2y^4$ became 0 because it contains no $x$ and is treated as a constant scalar).
      
        - Find $\frac{\partial f}{\partial y}$ (Treat $x$ as a constant number):$$\frac{\partial f}{\partial y} = 4x^3 \cdot (2)y + 0 + 8y^3 = 8x^3y + 8y^3$$
## The Gradient Vector ($\nabla f$):
  - The Gradient (denoted by the upside-down triangle symbol $\nabla$, called "nabla") is simply a vector that packs together all the individual partial derivatives of a multivariable function.If you have a function $f(x, y, z)$, its gradient vector is:$$\nabla f = \begin{bmatrix} \frac{\partial f}{\partial x} \\ \frac{\partial f}{\partial y} \\ \frac{\partial f}{\partial z} \end{bmatrix}$$
  
  - Geometric Property of the Gradient:
    - The gradient vector points in the direction of the steepest ascent on the function's surface. The magnitude (length) of the vector tells you how steep that incline is.
    
### AI Application:
  - Gradient Descent: In a neural network, we do not want to go uphill; we want to go downhill to find the lowest possible error.Because the gradient vector points directly uphill, if we take the negative of the gradient ($-\nabla J$), it points directly downhill toward the lowest error. This is the foundation of Gradient Descent.
  
  - The Mathematical Update Rule: For every weight ($w_i$) in your neural network, the optimization algorithm updates its value simultaneously using its corresponding slot in the gradient vector:$$w_i = w_i - \alpha \frac{\partial J}{\partial w_i}$$Where $\alpha$ (Alpha) is the Learning Rate—a scalar that controls how large of a step the model takes down the slope.

```python
import sympy as sp

# Define symbols for weights and bias
w1, w2, b = sp.symbols('w1 w2 b')

# Define a mock Mean Squared Error loss function
loss = (3*w1 + 5*w2 + b - 10)**2

# Compute the gradient components (partial derivatives)
grad_w1 = sp.diff(loss, w1)
grad_w2 = sp.diff(loss, w2)
grad_b  = sp.diff(loss, b)

# Output the full Gradient Vector
gradient_vector = [grad_w1, grad_w2, grad_b]
print(f"Gradient Vector:\n w1 component: {grad_w1}\n w2 component: {grad_w2}\n b component: {grad_b}")
```

### Directional Derivatives
- A directional derivative generalizes this concept: it computes the instantaneous rate of change (slope) when moving in any arbitrary direction specified by a vector.
- Mathematically: $$D_{\vec{u}}f(x, y) = \nabla f(x, y) \cdot \vec{u}$$Expanding the dot product gives:$$D_{\vec{u}}f(x, y) = \frac{\partial f}{\partial x}u_1 + \frac{\partial f}{\partial y}u_2$$
 Critical Constraint: The direction vector $\vec{u}$ must be normalized to a length of 1 before computing. If a raw vector $\vec{v}$ is provided, it must be converted first: $\vec{u} = \frac{\vec{v}}{\|\vec{v}\|_2}$.

 - Calculation Example: Let $f(x, y) = 2x^2 + 3xy$. Find the slope at the point $(1, 2)$ in the direction of the vector $\vec{v} = \begin{bmatrix} 3 \\ 4 \end{bmatrix}$.

 1. Compute the Gradient Vector $\nabla f$:$$\frac{\partial f}{\partial x} = 4x + 3y \implies \text{At } (1,2): 4(1) + 3(2) = 10$$$$\frac{\partial f}{\partial y} = 3x \implies \text{At } (1,2): 3(1) = 3$$$$\nabla f(1, 2) = \begin{bmatrix} 10 \\ 3 \end{bmatrix}$$

 2. Normalize the Direction Vector $\vec{v}$:$$\|\vec{v}\|_2 = \sqrt{3^2 + 4^2} = \sqrt{25} = 5$$$$\vec{u} = \frac{1}{5}\begin{bmatrix} 3 \\ 4 \end{bmatrix} = \begin{bmatrix} 0.6 \\ 0.8 \end{bmatrix}$$

 3. Compute the Directional Derivative via Dot Product:$$D_{\vec{u}}f(1, 2) = \nabla f(1, 2) \cdot \vec{u} = \begin{bmatrix} 10 \\ 3 \end{bmatrix} \cdot \begin{bmatrix} 0.6 \\ 0.8 \end{bmatrix}$$$$D_{\vec{u}}f(1, 2) = (10 \cdot 0.6) + (3 \cdot 0.8) = 6 + 2.4 = 8.4$$The instantaneous slope of the surface at $(1,2)$ heading toward direction $\vec{v}$ is exactly $8.4$.


### Operational Relevance in Machine Learning (Optimization)
- Standard Gradient Descent limits updates strictly along the direction of steepest descent ($-\nabla J$). However, advanced optimization algorithms use directional derivatives to evaluate alternative paths:
  - Line Search Methods: Once a descent direction is chosen, optimization algorithms compute directional derivatives along that specific vector line to determine how far parameters can shift before the loss function begins increasing again.
  - Conjugate Gradient Descent: Instead of stepping blindly, it calculates orthogonal directional trajectories to reach the minimum without undoing progress made in prior steps.

```python
import sympy as sp

# Define symbols
x, y = sp.symbols('x y')

# Define function and point
f = 2*x**2 + 3*x*y
point = {x: 1, y: 2}

# Compute gradient components at the specific point
grad_x = sp.diff(f, x).subs(point)
grad_y = sp.diff(f, y).subs(point)
grad_vector = sp.Matrix([grad_x, grad_y])

# Define and normalize direction vector v = [3, 4]
v = sp.Matrix([3, 4])
u = v / v.norm()

# Compute Directional Derivative via Dot Product
directional_derivative = grad_vector.dot(u)
print(f"Directional Derivative: {float(directional_derivative)}")  # Output: 8.4
```

### The Jacobian Matrix:
  - The Jacobian Matrix generalizes the first derivative by mapping a vector-valued function with multiple inputs to multiple outputs.
  - If you have a function $\mathbf{f}: \mathbb{R}^n \to \mathbb{R}^m$ that takes an $n$-dimensional vector and outputs an $m$-dimensional vector, the Jacobian $J$ is an $m \times n$ matrix containing all possible first-order partial derivatives.$$J = \begin{bmatrix}
\frac{\partial f_1}{\partial x_1} & \frac{\partial f_1}{\partial x_2} & \cdots & \frac{\partial f_1}{\partial x_n} \\
\frac{\partial f_2}{\partial x_1} & \frac{\partial f_2}{\partial x_2} & \cdots & \frac{\partial f_2}{\partial x_n} \\
\vdots & \vdots & \ddots & \vdots \\
\frac{\partial f_m}{\partial x_1} & \frac{\partial f_m}{\partial x_2} & \cdots & \frac{\partial f_m}{\partial x_n}
\end{bmatrix}$$

- Operational Example: Let $\mathbf{f}(x, y) = \begin{bmatrix} x^2y \\ x + 3y \end{bmatrix}$. Here, $f_1 = x^2y$ and $f_2 = x + 3y$.$$J = \begin{bmatrix}
\frac{\partial f_1}{\partial x} & \frac{\partial f_1}{\partial y} \\
\frac{\partial f_2}{\partial x} & \frac{\partial f_2}{\partial y}
\end{bmatrix} = \begin{bmatrix} 2xy & x^2 \\ 1 & 3 \end{bmatrix}$$

#### AI ApplicationsDeep Learning (Backpropagation):
  -  When layers pass vectors forward, the Chain Rule across vector-valued activations requires multiplying successive Jacobian matrices together.
  - Computer Vision: Used in optical flow and coordinate transformations (e.g., transforming camera space to pixel coordinates) to compute local spatial deformations.

```python
import sympy as sp

x, y, z = sp.symbols('x y z')

f_vector = sp.Matrix([x**2 * y, x + 3*y])
jacobian_matrix = f_vector.jacobian([x, y])
print("Jacobian Matrix:\n", jacobian_matrix)
```

### The Hessian Matrix:
- The Hessian Matrix is a square matrix of second-order partial derivatives of a scalar-valued function $f: \mathbb{R}^n \to \mathbb{R}$. It tracks the curvature (concavity) of a multivariable surface.
- For a function $f(x, y, z)$, the Hessian $H$ is a symmetric $n \times n$ matrix:$$H = \begin{bmatrix}
\frac{\partial^2 f}{\partial x^2} & \frac{\partial^2 f}{\partial x \partial y} & \frac{\partial^2 f}{\partial x \partial z} \\
\frac{\partial^2 f}{\partial y \partial x} & \frac{\partial^2 f}{\partial y^2} & \frac{\partial^2 f}{\partial y \partial z} \\
\frac{\partial^2 f}{\partial z \partial x} & \frac{\partial^2 f}{\partial z \partial y} & \frac{\partial^2 f}{\partial z^2}
\end{bmatrix}$$

-  Operational Example: Let $f(x, y) = x^3 - 2xy + y^2$.
  - First partials: $\frac{\partial f}{\partial x} = 3x^2 - 2y$, $\frac{\partial f}{\partial y} = -2x + 2y$
  - Second partials: $\frac{\partial^2 f}{\partial x^2} = 6x$, $\frac{\partial^2 f}{\partial y^2} = 2$, $\frac{\partial^2 f}{\partial x \partial y} = \frac{\partial^2 f}{\partial y \partial x} = -2$ $$H = \begin{bmatrix} 6x & -2 \\ -2 & 2 \end{bmatrix}$$
  
#### AI Applications:
  - Second-Order Optimization: Algorithms like Newton's Method use $H^{-1}$ alongside the gradient to alter step sizes based on local curvature, accelerating convergence.
  - Saddle Point Identification: A critical point ($\nabla f = 0$) is evaluated using the eigenvalues of $H$. If $H$ is positive definite, it is a local minimum. If the eigenvalues have mixed signs, it is a saddle point.

```python
x, y, z = sp.symbols('x y z')

f_scalar = x**3 - 2*x*y + y**2
hessian_matrix = sp.hessian(f_scalar, [x, y])
print("\nHessian Matrix:\n", hessian_matrix)
```

### Taylor Series (Multivariable Approximation)
- The Taylor Series approximates a complex, non-linear function near a specific point using a polynomial series of its derivatives.
- The second-order multivariable Taylor expansion of a scalar function $f(\mathbf{x})$ around a point $\mathbf{a}$ is written as:$$f(\mathbf{x}) \approx f(\mathbf{a}) + \nabla f(\mathbf{a})^T (\mathbf{x} - \mathbf{a}) + \frac{1}{2} (\mathbf{x} - \mathbf{a})^T H(\mathbf{a}) (\mathbf{x} - \mathbf{a})$$

#### AI Applications:
  - Optimization Proofs: Used to prove that gradient descent updates lower the loss function locally (via the First-Order Taylor Theorem).
  - Function Approximations: Used to simplify highly non-linear loss functions into local quadratic forms for faster calculation.

### Integrals & Multiple Integrals:
- An integral computes the net area under a curve. A multiple integral extends this to higher dimensions, calculating volume or hyper-volume over a region.$$\iint_R f(x, y) \, dA = \int_a^b \int_c^d f(x, y) \, dy \, dx$$

#### AI Applications:
  - Probability Theory: Continuous random variables require integration to find probabilities. The total area under a Probability Density Function (PDF) must equal 1:$$\int_{-\infty}^{\infty} p(x) \, dx = 1$$
  - Joint Distributions: Multiple integrals compute probabilities over multiple dependent random variables simultaneously: $P(A \cap B) = \iint_R p(x, y) \, dx \, dy$.

### Vector Calculus: Divergence & Curl
- Vector calculus views space as a Vector Field, where every single point in coordinate space is assigned a vector (e.g., fluid velocity or gravitational pull).
1.  Divergence ($\text{div }\mathbf{F} \text{ or } \nabla \cdot \mathbf{F}$):
     - Divergence calculates the net flow of a vector field out of or into a single point. It outputs a scalar value.$$\nabla \cdot \mathbf{F} = \frac{\partial F_x}{\partial x} + \frac{\partial F_y}{\partial y} + \frac{\partial F_z}{\partial z}$$
     - $\nabla \cdot \mathbf{F} > 0$: The point acts as a Source (fluid expanding outwards).
     - $\nabla \cdot \mathbf{F} < 0$: The point acts as a Sink (fluid condensing inwards).
2. Curl ($\text{curl }\mathbf{F} \text{ or } \nabla \times \mathbf{F}$):
    - Curl measures the infinitesimal rotation of a vector field around a given point. It outputs a vector indicating the axis of rotation.$$\nabla \times \mathbf{F} = \begin{vmatrix} \mathbf{i} & \mathbf{j} & \mathbf{k} \\ \frac{\partial}{\partial x} & \frac{\partial}{\partial y} & \frac{\partial}{\partial z} \\ F_x & F_y & F_z \end{vmatrix}$$
    - AI Relevancy: Rare in traditional ML. It appears primarily in physics-informed neural networks (PINNs) modeling fluid dynamics, aerodynamics, or electromagnetism.

```python
field = [x**2, y**2, z**2]
divergence = sp.diff(field[0], x) + sp.diff(field[1], y) + sp.diff(field[2], z)
print("\nDivergence of field:", divergence)
```
